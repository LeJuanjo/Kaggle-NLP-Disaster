# I/ Pre-Processing

-> Remove site links

```python
train['text'] = train['text'].str.replace('http[^\s]*',"")
test['text'] = test['text'].str.replace('http[^\s]*',"")
```

-> Remove special characters

```python
train['txt'] = train['text'].apply(lambda x : re.sub("[^A-Z]", " ", str(x),0,re.IGNORECASE))
test['txt'] = test['text'].apply(lambda x : re.sub("[^A-Z]", " ", str(x),0,re.IGNORECASE))
```

-> Convert to string

```python
train['txt'] = train['txt'].astype('string')
test['txt'] = test['txt'].astype('string')
```

```python
train['txt']
```

-> Remove stopwords

```python
stop_words = stopwords.words('english')
train['txt'] = train['txt'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
test['txt'] = test['txt'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
```

-> Convert to Lower Case

```python
train['txt'] = train['txt'].apply(lambda x: x.lower())
test['txt'] = test['txt'].apply(lambda x: x.lower())
```

-> Remove Strings with 1 letter

```python
train['txt'] =  train['txt'].apply(lambda x: ' '.join( [w for w in x.split() if len(w)>1] ))
test['txt'] =  test['txt'].apply(lambda x: ' '.join( [w for w in x.split() if len(w)>1] ))
```

```python
train['txt']
```

-> Tokenisation & Lemmatisation

```python
w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()

def lemmatize_text(text):
    return [lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(text)]

train['txt'] = train['txt'].apply(lemmatize_text)
test['txt'] = test['txt'].apply(lemmatize_text)
```

```python
train['txt'] = train['txt'].apply(lambda x: ' '.join( [w for w in x]))
test['txt'] = test['txt'].apply(lambda x: ' '.join( [w for w in x]))
```

```python
train['txt']
```

# II/ Applying Algorithms

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
vocab_size = 10000
embedding_dim = 16
max_length = 100
trunc_type='post'
padding_type='post'
oov_tok = "<OOV>"
training_size = 20000
```

```python
tokenizer = Tokenizer(num_words = vocab_size, oov_token= oov_tok) #the OOV allows for new words to be tokenized that were not pressent on the training data
tokenizer.fit_on_texts(train["txt"])
word_index = tokenizer.word_index
print(word_index)
```

```python
train['token'] = tokenizer.texts_to_sequences(train['txt'])
test['token'] = tokenizer.texts_to_sequences(test['txt'])
print (test['token'])
```

```python
training_sentences = pad_sequences(train['token'],  maxlen=max_length, padding=padding_type, truncating=trunc_type)
test_sentences = pad_sequences(test['token'], maxlen=max_length, padding=padding_type, truncating=trunc_type)
print(training_sentences)
```

```python
training_label = np.array(train["target"])
print(training_label)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim, input_length=max_length),
    tf.keras.layers.SimpleRNN(units= 24),
    tf.keras.layers.Dense(24, activation='relu'),
    tf.keras.layers.Dense(24, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
num_epochs = 30
model.fit(training_sentences, training_label, epochs=num_epochs,validation_split=0.2, verbose=2)
```

```python
test_pred = model.predict(test_sentences)
print(test_pred[0])
```

```python
from keras.models import Sequential
from keras.layers import Input, LSTM, Dense
from keras.layers.embeddings import Embedding
import keras.backend as K
```

```python
model =tf.keras. Sequential()
model.add(Embedding(vocab_size, 32, input_length=max_length))
# While passing output to next LSTM layer set return_sequence to Ture.
model.add(LSTM(32, return_sequences=True)) 
model.add(LSTM(32))
model.add(Dense(32,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss=['binary_crossentropy'] , optimizer='adam', metrics=['accuracy'])
```

```python
scores = model_selection.cross_val_score(clf, train_vectors, train["target"], cv=3, scoring="f1")
scores
```
